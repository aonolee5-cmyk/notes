```python
# 缺失值的处理
import numpy as np
import pandas as pd
s=pd.Series([1,2,np.nan,None,pd.NA])
df=pd.DataFrame([[1,pd.NA,2],[2,3,5],[None,4,6]])
print(s)
print('='*20)


# 查看是否是缺失值
print(s.isna()) # isna()
print('='*20)
print(s.isnull()) # isnull()
print('='*20)
print('缺失值个数：',s.isna().sum())
print('='*20)
print(df.isna()) # 与series一致
print('='*20)
print(df.isna().sum())


# 
```

```python
# 剔除缺失值
print(s.dropna()) # 只剔除缺失值 Series
print('='*20)
print(df.dropna()) # 若有缺失值，直接删除整条数据 DataFrame
print('='*20)
print(df.dropna(how='all')) # 如果所有的值都为缺失值，删除这一行
print('='*20)
print(df.dropna(thresh=2)) # 与参数how不同，在dropna()方法中是看缺失值，而thresh是看有效值。比如thresh=2，只要有效值大于等于这个数，保留数据。相反则删除。同时参数 how 和 thresh 两者不能共存
```

```python
# 填充缺失值
df=pd.read_csv('data/weather_withna.csv')
df.tail()
print(df.isna().sum()) # 找出缺失值总数
print('='*20)


print(df.fillna({'temp_max':25,'temp_min':16,'wind':4}).tail()) # 使用字典填充 # fillna()
print('='*20)


print(df.fillna(df[['wind','temp_max','temp_min']].mean()).tail()) # 使用统计值填充 # fillna()
print('='*20)


print(df.ffill()) # 前向填充(forward fill) df.ffill() 会将缺失值用前一个非缺失值填充，沿着列方向（默认）进行。
print('='*20)


print(df.bfill()) # 后向填充(backward fill) df.bfill() 用下一个非缺失值把 NaN 填掉，默认沿着列方向（axis=0）进行。
print('='*20)


```

## 分组聚合

```python
# df.groupby('分组的字段')['聚合的字段'].聚合函数
import pandas as pd
df=pd.read_csv('data/employees.csv')
print(df['department_id'].isna().sum()) # 找出缺失值总数
df=df.dropna(subset=['department_id']) # 由于缺失值只有一个，选择直接删除，对结果影响不大
df['department_id']=df['department_id'].astype('int64')
df
```

```python
# 计算不同部门的平均薪资
df.groupby('department_id').groups # 查看分组的情况 字典类型 索引号
df.groupby('department_id').get_group(110) # 查看具体某个分组数据
```

```python
df2=df.groupby('department_id')[['salary']].mean().round(2)
df2.sort_values('salary',ascending=False)
```

```python
# 计算不同部门不同岗位的平均工资
df2=df.groupby(['department_id','job_id'])[['salary']].mean().round(2)
df2=df2.reset_index()
df2.sort_values('salary',ascending=False)
```

## 企鹅数据分析

```python
# 1.导入相关库

import pandas as pd
import numpy as np
# 2.导入数据

df=pd.read_csv('data/penguins.csv')
df.head()
```

```python
# 3.数据清洗 缺失值，重复值，数据类型的转换
print(df.isna().sum()) # 查看缺失值总数
df.dropna(inplace=True) # 去除缺失值，并直接在原表上操作
```

```python
# 4.数据特征的构造
df['sex']=df['sex'].astype('category')
df['bill_ratio']=(df['bill_length_mm']/df['bill_depth_mm']).round(3) # 构建一个比值
df
```

```python
# 5.数据分析
# 数据分箱 把体重分为三个等级
df['体重等级']=pd.cut(df['body_mass_g'],3,labels=['低','中','高'])
df['体重等级'].value_counts()
```

```python
# 按性别分析
df.groupby(['sex','island'],observed=False).agg({
    'body_mass_g':['mean','count']
})
```

## 睡眠质量分析

```python
import pandas as pd
import numpy as np

df=pd.read_csv('data/sleep.csv')
df.head()
```

```python
df.drop(columns='sleep_disorder',inplace=True)
df.head()
```

```python
# 数据特征的构造
df['gender']=df['gender'].astype('category')
df['occupation']=df['occupation'].astype('category')
df['bmi_category']=df['bmi_category'].astype('category')
df[['high','low']]=df['blood_pressure'].str.split('/',expand=True)

#  对睡眠质量进行分箱
df['quality_level']=pd.cut(df['sleep_quality'],
                           3,
                           labels=['差','良','优'])


# 对年龄层次分箱
df['age_level']=pd.cut(df['age'],
                       3,
                       labels=['青年','中年','老年'])

df
```

```python
# 数据的统计
# 根据不同的bmi分组
df.groupby(['age_level','bmi_category'],observed=True).agg({
    'sleep_duration':'mean',
    'sleep_quality':'mean',
    'stress_level':'mean'
}).round(3)
```
