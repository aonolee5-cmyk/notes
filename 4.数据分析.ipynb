```python
# 缺失值的处理
import numpy as np
import pandas as pd
s=pd.Series([1,2,np.nan,None,pd.NA])
df=pd.DataFrame([[1,pd.NA,2],[2,3,5],[None,4,6]])
print(s)
print('='*20)


# 查看是否是缺失值
print(s.isna()) # isna()
print('='*20)
print(s.isnull()) # isnull()
print('='*20)
print('缺失值个数：',s.isna().sum())
print('='*20)
print(df.isna()) # 与series一致
print('='*20)
print(df.isna().sum())


# 
```

```python
# 剔除缺失值
print(s.dropna()) # 只剔除缺失值 Series
print('='*20)
print(df.dropna()) # 若有缺失值，直接删除整条数据 DataFrame
print('='*20)
print(df.dropna(how='all')) # 如果所有的值都为缺失值，删除这一行
print('='*20)
print(df.dropna(thresh=2)) # 与参数how不同，在dropna()方法中是看缺失值，而thresh是看有效值。比如thresh=2，只要有效值大于等于这个数，保留数据。相反则删除。同时参数 how 和 thresh 两者不能共存
```

```python
# 填充缺失值
df=pd.read_csv('data/weather_withna.csv')
df.tail()
print(df.isna().sum()) # 找出缺失值总数
print('='*20)


print(df.fillna({'temp_max':25,'temp_min':16,'wind':4}).tail()) # 使用字典填充 # fillna()
print('='*20)


print(df.fillna(df[['wind','temp_max','temp_min']].mean()).tail()) # 使用统计值填充 # fillna()
print('='*20)


print(df.ffill()) # 前向填充(forward fill) df.ffill() 会将缺失值用前一个非缺失值填充，沿着列方向（默认）进行。
print('='*20)


print(df.bfill()) # 后向填充(backward fill) df.bfill() 用下一个非缺失值把 NaN 填掉，默认沿着列方向（axis=0）进行。
print('='*20)


```

## 数据变形

```python
data={
    'ID':[1,2],
    'name':['alice','bob'],
    'Math':[98,85],
    'English':[88,92],
    'Science':[95,89]
}
df=pd.DataFrame(data)
df.T # 行列转置
```

```python
# 宽表转化为长表
df2=pd.melt(df,id_vars=['ID','name'],var_name='科目',value_name='分数')
df2.sort_values('name')
```

```python
# 长表转宽表
pd.pivot(df2,index=['ID','name'],columns='科目',values='分数')
```

## 数据分箱 pd.cut(x,bins,labels)

```python
import pandas as pd
df=pd.read_csv('data/employees.csv')
df.head(10)
```

```python
df1=df.head(10)[['employee_id','salary']]
pd.cut(df1['salary'],bins=2) # bins 自动等宽分 起始值，结束值分别时所有值的最小值和最大值
pd.cut(df1['salary'],bins=3).value_counts()
```

```python
pd.cut(df1['salary'],bins=[0,10000,20000,30000]) # bins用列表自定义分割
```

```python
pd.cut(df1['salary'],bins=[0,10000,20000,30000]).value_counts()
```

```python
df1['收入水平']=pd.cut(df1['salary'],
                   bins=[0,10000,20000,30000],
                   labels=['低','中','高'])
df1
```

```python
pd.qcut(df1['salary'],3).value_counts() # 等频或分位数划分
```

```python
# 睡眠数据
df=pd.read_csv('data/sleep.csv')
df.head(10)
```
