#### 1.绘制折线图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams # 字体
rcParams['font.family']='SimHei'

# 创建图表，设置大小
plt.figure(figsize=(10,5))

month=['一月','二月','三月','四月']
sales=[100,40,200,140]

# 绘制折线图
plt.plot(month,sales,
         label='产品A',
         color='red',
         linewidth=3,
         marker='o'
         )
plt.title('2025年销售趋势',color='red',fontsize=20) # 设置标题属性
plt.xlabel('月份',fontsize=20) # 添加坐标轴标签
plt.ylabel('销售额',fontsize=20)

# 添加图例
plt.legend(loc='upper left')

# 添加网格线
plt.grid(axis='y',alpha=1)


# 设置刻度大小
plt.xticks(rotation=0,fontsize=15)
plt.yticks(rotation=0,fontsize=15)

# 设置y轴的范围
plt.ylim(0,250)

# 在每个数据点上显示数值
for x,y in zip(month,sales):
    print(x,y)
    plt.text(x,y,str(y),ha='center',va='bottom',fontsize=20)


# 显示图表
plt.show()
```

#### 2.柱状图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family']='SimHei'

plt.figure(figsize=(10,5)) # 图表大小

# 数据
subjects = ['语文','数学','英语','物理']
scores = [90,59,79,40]


# 绘制柱状图 bar
plt.bar(subjects,scores,
        label='xx',
        width=0.4)


# 添加标题
plt.title('2025年期末成绩',color='red',fontsize=18)
plt.xlabel('科目',fontsize=16)
plt.ylabel('分数',fontsize=15)

# 设置刻度字体大小
plt.xticks(rotation=0,fontsize=18)
plt.yticks(rotation=0,fontsize=18)

# 设置y轴范围
plt.ylim(0,100)

# 添加图例
plt.legend(loc='upper right')

for x,y in zip(subjects,scores):
    plt.text(x,y,str(y),ha='center',va='bottom',fontsize=18)

# 自动优化排版
plt.tight_layout()

plt.show()
```

#### 条形图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family']='SimHei'

# 数据
subjects = ['语文','数学','英语','物理']
scores = [90,59,79,40]

# 图的大小
plt.figure(figsize=(10,5))

# x轴的范围
plt.xlim(0,100)

# 绘制条形图
plt.barh(subjects,scores,label='xx') # 条形图 barh

plt.title('2025年期末成绩',fontsize=18,color='red')

# 添加x,y轴描述
plt.xlabel('分数',fontsize=17)
plt.ylabel('科目',fontsize=17)


# 设置刻度大小
plt.xticks(rotation=0,fontsize=18)
plt.yticks(rotation=0,fontsize=18)

# 添加图例
plt.legend(loc='upper right')


# 自动优化排版
plt.tight_layout()

plt.show()
```

#### 3.饼图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family']='SimHei'

plt.figure(figsize=(10,5))


# 数据
things=['学习','娱乐','运动','睡觉','其他']
times=[5,2,4,9,4]

# 自定义配色
color=[]

# 添加标题
plt.title('一天的时间',color='r',fontsize=15)

plt.pie(times,labels=things,
        autopct='%.1f%%',
        startangle=15) # autopct 显示百分比

# 自动优化排版
plt.tight_layout()

plt.show()
```

#### 环形图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family']='SimHei'

plt.figure(figsize=(10,5))


# 数据
things=['学习','娱乐','运动','睡觉','其他']
times=[5,2,4,9,4]

# 自定义配色
color=[]

# 添加标题
plt.title('一天的时间',color='r',fontsize=15)

plt.pie(times,labels=things,
        autopct='%.1f%%',
        startangle=15,
        wedgeprops={'width':0.6},
        pctdistance=0.7) # autopct 显示百分比；wedgeprops 环形图；pctdistance 设置百分比显示位置

plt.text(0,0,'总计：\n100%',ha='center',va='bottom',fontsize=16)

# 自动优化排版
plt.tight_layout()

plt.show()
```

#### 爆炸式饼图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family']='SimHei'

plt.figure(figsize=(10,5))


# 数据
things=['学习','娱乐','运动','睡觉','其他']
times=[5,2,4,9,4]

# 自定义配色
color=[]

# 显示突出的数据
explode=[0.1,0,0,0,0]

# 添加标题
plt.title('一天的时间',color='r',fontsize=15)

plt.pie(times,labels=things,
        autopct='%.1f%%',
        startangle=15,
        explode=explode,
        shadow=True # 添加阴影
        ) # autopct 显示百分比

# 自动优化排版
plt.tight_layout()

plt.show()
```

#### 4.散点图

```python
import matplotlib.pyplot as plt
import random
from matplotlib import rcParams
rcParams['font.family']='SimHei'

# 数据
X=[]
Y=[]
for i in range(1000):
    tmp=random.uniform(0,10)
    X.append(tmp)
    Y.append(2*tmp + random.gauss(0,2))


# 设置图表基本参数
plt.figure(figsize=(10,5))

# 设置刻度大小
plt.xticks(rotation=0,fontsize=18)
plt.yticks(rotation=0,fontsize=18)

# 设置标题
plt.title('X与Y的关系',color='r',fontsize=18)

plt.xlabel('X变量',color='r',fontsize=18)
plt.ylabel('Y变量',color='r',fontsize=18)

plt.ylim(0,30)
plt.xlim(0,10)

plt.scatter(X,Y,
            color='b',
            alpha=0.8, # 设置透明度
            s=20, # 设置大小
            label='数据'
            )

# 添加图例
plt.legend(loc=('upper left'))

# plt.tight_layout()

plt.plot([0,10],[0,20],color='r',linewidth=3)

plt.show()
```

#### 5.箱线图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family']='SimHei'

data={
    '语文':[82,85,88,70,90,76,84,83,95],
    '数学':[75,80,79,93,88,82,87,89,92],
    '英语':[70,72,68,78,80,85,90,95]
}

plt.figure(figsize=(8,6))
plt.title('各科成绩分布(箱线图)')
plt.ylabel('分数')
plt.grid(True,alpha=0.5,linestyle='--',axis='y')

plt.boxplot(data.values(),tick_labels=data.keys())



plt.show()
```

```python
'''
折线图：趋势随时间变化
条形图/柱状图：类别之间对比
饼图：整体组成比例
散点图：两变量之间的相关性
箱线图：数据分布、异常
'''
```

## 温度分析

```python
from cProfile import label
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family']='Microsoft YaHei'

df=pd.read_csv('data/weather.csv')
df['date']=pd.to_datetime(df['date']) # 处理数据
df=df[df['date'].dt.year==2015]

# 绘制气温的趋势变化图
plt.figure(figsize=(15,6))
plt.title('2015年气温的趋势变化图',color='r',fontsize=18)
plt.plot(df['date'],df['temp_max'],label='最高气温',linewidth=2)
plt.plot(df['date'],df['temp_min'],label='最低气温',linewidth=2)

# 设置坐标轴标签
plt.xlabel('时间',rotation=0,fontsize=18)
plt.ylabel('温度',rotation=0,fontsize=18)

plt.xticks(fontsize=18)
plt.yticks(fontsize=18)


plt.legend(loc=('upper right'))

plt.show()
```

```python
df['temp_mean']=(df['temp_max']+df['temp_min'])/2
# 绘制平均气温的趋势变化图
plt.figure(figsize=(15,6))
plt.title('2015年平均气温的趋势变化图',color='r',fontsize=18)
plt.plot(df['date'],df['temp_mean'],label='平均气温')
# 设置坐标轴标签
plt.xlabel('时间',rotation=0,fontsize=18)
plt.ylabel('温度',rotation=0,fontsize=18)

plt.xticks(fontsize=18)
plt.yticks(fontsize=18)


plt.legend(loc=('upper right'))

plt.show()
```

### seaborn学习

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family']='Microsoft YaHei'

penguins=pd.read_csv('data/penguins.csv')
penguins.dropna(inplace=True)
penguins.head()
```

```python
# 绘制直方图
sns.histplot(data=penguins,x='species')
```

```python
# 绘制喙长度的核密度估计图
sns.kdeplot(data=penguins,x='bill_length_mm')
```

```python
sns.histplot(data=penguins,x='bill_length_mm',kde=True)
```

```python
'''
计数图 用于绘制分类变量的计数分布图，显示每个类别在数据集中出现的次数，是分析分类数据非常
直观的工具，可以快速了解类别的分布情况
'''
```

```python
# 绘制不同岛屿企鹅数量的计数图
sns.countplot(data=penguins,x='island')
```

```python
# 绘制横轴为体重，纵轴为脚蹼长度的散点图，可通过hue参数设置不同组别进行对比
sns.scatterplot(data=penguins,
                x='body_mass_g',
                y='flipper_length_mm',
                hue='sex')
```

```python
# 蜂窝图
# 通过jointplot()函数，设置kind='hex'来绘制蜂窝图
sns.jointplot(data=penguins,
              x='body_mass_g',
              y='flipper_length_mm',
              kind='hex')
```

```python
# 二维核密度估计图
# 通过kdeplot()函数，同时设置x参数和y参数来绘制二维核密度估计图
# 通过fill=True设置为填充，通过cbar=True设置显示颜色示意条
sns.kdeplot(data=penguins,
              x='body_mass_g',
              y='flipper_length_mm',
              fill=True,
              cbar=True)
```

```python
#条形图
sns.barplot(data=penguins,
            x='species',y='bill_length_mm',
            estimator='mean',
            errorbar=None)
```

```python
# 箱线图
sns.boxplot(data=penguins,x='species',y='bill_length_mm')
```

```python
# 小提琴图
sns.violinplot(data=penguins,x='species',y='bill_length_mm')
```

```python
# 成对关系图
sns.pairplot(data=penguins,hue='species')
```
