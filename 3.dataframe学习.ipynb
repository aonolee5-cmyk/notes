## eg1:学生成绩分析
### 某班级的学生成绩如下
### 1.计算每位学生的总分和平均分
### 2.找出数学成绩高于90分或英语成绩高于85的学生
### 3.按总分从高到低排序，并输出前3名学生

```python
import pandas as pd
data={
    '姓名':['张三','李四','王五','赵六','钱七'],
    '数学':[85,92,78,88,95],
    '英语':[90,88,85,92,80],
    '物理':[75,80,88,85,90]
}
data
```

```python
scores=pd.DataFrame(data)
scores
```

```python
# 总分和平均分
scores['总分']=scores[['数学','英语','物理']].sum(axis=1)
scores['平均分']=scores[['数学','英语','物理']].mean(axis=1).round(4)
scores
```

```python
# 2.找出数学成绩高于90分或英语成绩高于85的学生
scores[(scores['数学']>90)|(scores['英语']>85)]
```

```python
# 3.按总分从高到低排序，并输出前3名学生
r1=scores.nlargest(3,'总分')
r1
# r2=scores.sort_values('总分',ascending=False).head(3)
# r2
```

## eg2:销售数据分析
### 某公司销售数据如下
### 1.计算每种产品的总销售额（销售额=单价X销量）
### 2.找出销售额最高的产品
### 3.按销售额从高到低排序，并输出所有产品信息

```python
import pandas as pd
data={
    '产品名称':['A','B','C','D'],
    '单价':[100,150,200,120],
    '销量':[20,30,20,40]
}
```

```python
df=pd.DataFrame(data)
df
```

```python
# 1.计算每种产品的总销售额（销售额=单价X销量）
df['销售额']=df['单价']*df['销量']
df
```

```python
# 2.找出销售额最高的产品
df.nlargest(1,'销售额')
```

```python
# 3.按销售额从高到低排序，并输出所有产品信息
df.sort_values('销售额',ascending=False)
```

## eg3:电商用户行为分析
### 某电商平台的用户行为数据如下
### 1.计算每位用户的总消费金额（消费金额=商品单价*购买数量）
### 2.找出消费金额最高的用户，并输出其所有信息
### 3.计算所有用户的平均消费金额（保留两位小数）
### 4.统计电子产品的总购买数量

```python
data={
    '用户id':[101,102,103,104,105],
    '用户名':['jk','qq','pt','xx','smz'],
    '商品类别':['电子产品','服饰','电子产品','家居','服饰'],
    '单价':[1200,300,800,150,200],
    '购买数量':[1,3,2,5,4]
}
```

```python
df=pd.DataFrame(data)
df
```

```python
# 1.计算每位用户的总消费金额（消费金额=商品单价*购买数量）
df['总消费金额']=df['单价']*df['购买数量']
df
```

```python
# 2.找出消费金额最高的用户，并输出其所有信息
df.nlargest(1,'总消费金额')
```

```python
# 3.计算所有用户的平均消费金额（保留两位小数）
df['平均消费']=df['总消费金额'].mean().round(2)
df
```

```python
# 4.统计电子产品的总购买数量
df[df['商品类别']=='电子产品']['购买数量'].sum()
```
